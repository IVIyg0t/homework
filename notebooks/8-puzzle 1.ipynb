```python
#Definitions

#Convert a 0-8 index into a row or column number
def row(position):
    return floor(position / 3)

def col(position):
    return position % 3

def index(row, col):
    return row*3 + col

def print_puzzle(state):
    print(f"""
    {state[0]} {state[1]} {state[2]}
    {state[3]} {state[4]} {state[5]}
    {state[6]} {state[7]} {state[8]}
    """)
        
def all_moves(state):
    space = state.index(0)
    r = row(space)
    c = col(space)
    moves = []
    if r == 0:
        moves.append("S")
    elif r == 1:
        moves += ["N", "S"]
    else:
        moves.append("N")
    if c == 0:
        moves.append("E")
    elif c == 1 :
        moves += ["E", "W"]
    else:
        moves.append("W")
    return moves

def do_move(state, move, prnt = False):
    new_state = list(state)
    space = new_state.index(0)
    r = row(space)
    c = col(space)
    if move == "N":
        move_index = index(r-1,c)
    elif move == "S":
        move_index = index(r+1,c)
    elif move == "E":
        move_index = index(r, c+1)
    elif move == "W":
        move_index = index(r, c-1)
    
    if prnt:
        print(f'row = {r} | col = {c} | move_index = {move_index}')

    new_state[space], new_state[move_index] = state[move_index], state[space]
    return new_state

def undo_move(state, move):
    print('Undo Move')
    if move == "N":
        return do_move(state,"S")
    elif move == "S":
        return do_move(state, "N")
    elif move == "E":
        return do_move(state, "W")
    elif move == "W":
        return do_move(state, "E")
    
def shuffle(state, s):
    new_state = list(state)
    for i in range(s):
        moves = all_moves(state)
        rand = choice(all_moves(state))
        st = do_move(new_state, rand)
    return st

def energy(puzzle, solved_puzzle):
    energy = 0
    for i, val in enumerate(puzzle):
        if val != 0:
            solved_i = solved_puzzle.index(val)
            distance = abs(i - solved_i)
            energy += distance
    return energy

def solved(state, solved_puzzle):
    return energy(state, solved_puzzle) == 0

#def energy(state):

    
#def solved(state):
    

```

```python
def simulated_anneal(initial_state, solved_puzzle, max_moves, p = 1):
    state = initial_state
    T = max_moves
    prev_energy = energy(state, solved_puzzle)
    while not solved(state, solved_puzzle) and T > 0:

        moves = all_moves(state)
        accept = False
        while not accept and not solved(state, solved_puzzle):
            accept=False
            rand = choice(moves)
            #print("Previou State")
            #print_puzzle(state)
            state = do_move(state, rand, prnt=False)
            #print("New State")
            #print_puzzle(state)
            curr_energy = energy(state, solved_puzzle)
            change_energy = curr_energy - prev_energy
            
            #print(f"Change Energy = {change_energy}")
            if change_energy <= 0:
                accept = True
            else:
                boltzmann = exp(-float(p*change_energy)/T)
                n = random()
                if n <= boltzmann:
                    accept = True
                if not accept:
                    undo_move(state, rand)
                prev_energy = curr_energy
            if not accept:
                print(accept)
                #T = T - 1
                #print("Move to", rand)
                #print_puzzle(state)
            
```
